```python
import torch
import torch.nn as nn
import numpy as np
import csv
import matplotlib.pyplot as plt
from torch.autograd import Variable
```

检查GPU信息

```python
torch.cuda.is_available(), torch.cuda.device_count(), torch.cuda.get_device_name(0)
```

```python
# Hyper Parameters
input_size = 78
output_size = 1
num_epochs = 10000
learning_rate = 0.000001

# Toy Dataset 

train_y = []
train_x = []
#train_x = np.loadtxt('./new_shangshi_arr.txt', dtype='float32', delimiter=',')
```

load特征训练集

```python
with open('./newshangshi_arr.txt', 'r', encoding = 'utf-8') as f:
    reader = csv.reader(f)
    for line in reader:
#         for i in range(len(line)):
#             line[i] = float(line[i].replace("e+", ""))
        train_x.append(line)
train_x = np.array(train_x, dtype = 'float32')
train_x.shape, train_x
```

load收入训练集

```python
with open('./2019shouru.csv', 'r', encoding = 'utf-8') as f:
    reader = csv.reader(f)
    for line in reader:
        train_y.append(line)
train_y = np.array(train_y, dtype = 'float32')
train_y = train_y.reshape(-1, 1)
#print (y_train.shape)
train_y.shape, train_y
```

```python
# Linear Regression Model
class LinearRegression(nn.Module):
    def __init__(self, input_size, output_size, hidden_size):
        super(LinearRegression, self).__init__()
        self.linear1 = nn.Linear(input_size, hidden_size)  
        self.linear2 = nn.Linear(hidden_size, output_size)
    
    def forward(self, x):
        #print ("******: ", x)
        out = self.linear1(x)
        out = self.linear2(out)
        return out

model = LinearRegression(input_size, output_size, 78)


# Loss and Optimizer
criterion = nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)  

# 将模型传入到GPU中
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
torch.cuda.set_device(1)
model.to(device)
```

```python
# Train the Model 
for epoch in range(num_epochs):
    # Convert numpy array to torch Variable
    inputs = Variable(torch.from_numpy(train_x)).to(device)
    targets = Variable(torch.from_numpy(train_y)).to(device)

    #print (targets)

    # Forward + Backward + Optimize
    optimizer.zero_grad()  
    outputs = model(inputs)

    #print("^^^^^^: ", outputs)

    loss = criterion(outputs, targets)
    loss.backward()
    optimizer.step()
    
    if (epoch+1) % 100 == 0:
        print ('Epoch [%d/%d], Loss: %.4f' 
               %(epoch+1, num_epochs, loss.item()))
```
